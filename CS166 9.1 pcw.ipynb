```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
lst_results = []
#strategy: 2 heads then 1 tails, bet 10 always
for i in range(1000):
    ini = 250 #have 250 dollars to start
    for j in range(20):
        choice = 1 #initiate the choice as head
        if j % 3 == 0:
            choice = 0 #if it's the 3knth flip, go with tail
        else:
            pass #go with head
        result = 0 #initiate the result
        flip = random.random()
        if flip < 0.6:
            result = 1 #head
        else:
            result = 0 #tail
        if choice == result:
            ini += 10 #win
        elif choice != result and ini >= 10:
            ini -= 10 #lose
        else:
            ini = 0 #bankrupt
            break
    lst_results.append(ini)
```

```python
print("Confidence intervals are", np.percentile(lst_results, [2.5, 97.5]))
print("Mean is", np.mean(lst_results))
print("Mean return is", np.mean(lst_results)/250 - 1)
```
